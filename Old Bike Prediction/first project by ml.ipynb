```python
brand_dict = {brand:i for i, brand in enumerate(list(df2['brand'].value_counts().head(16).keys()),start = 1)}
print(brand_dict)
df2['brand']=df2['brand'].map(brand_dict)
```

```python
df2.drop(['bike_name','city'],axis=1,inplace=True)
```

```python
df2
```

```python
x = df2.drop('price',axis=1)
y = df2[['price']]
```

```python
y
```

```python
x
```

```python
# Train Test Split model
from sklearn.model_selection import train_test_split
#scikit-learn
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)  # 80% training size
```

```python
from sklearn.linear_model import LinearRegression  # Algorithm import 
```

```python
lnr = LinearRegression()
```

```python
lnr.fit(x_train,y_train)
```

```python
x_train.shape
```

```python
x_test
```

```python
x_train.isnull().sum()
```

```python
y_test
```

```python
# Seen data
lnr.score(x_train,y_train)* 100   # 74% Scored   , Training Score
```

```python
# Unseen Data
lnr.score(x_test,y_test)* 100  # Test data
```

```python
# prediction


prediction = lnr.predict(x_test)
```
