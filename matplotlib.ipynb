```python
ls1 = [45,50,55,60,35]
ls2 = [30,60,90,100,120]
plt .scatter(x=ls1,y=ls2,marker='D',s=450,color='black')   # s= size, marker = symbol,D= diamond
plt.ylabel('list - 2 ')
plt.xlabel('list -1 ')
plt.title("scatter plot")
plt.show()
```
